```python
Age_dict={'20-25':1, 'Less than 20':2, '45 and above':3, '30-35':4, '35-44':5, '25-30':5}
Age_dict
```

```python
df['Age']= df['Age'].map(Age_dict)
```

```python
df.dtypes
```

```python
cat_columns = df.select_dtypes(include='object').columns

for col in cat_columns:
    unique_vals = df[col].unique()
    mapping_dict = {}
    i=1
    for val in unique_vals:
        mapping_dict[val]=i
        i+=1
    df[col] = df[col].map(mapping_dict)
```

```python
df.to_csv("clean_pcos_data.csv")
```
