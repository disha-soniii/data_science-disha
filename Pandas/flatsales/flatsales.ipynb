```python
areaWithType_dict={'Carpet Area':1, 'Super Area':2, 'Plot Area':3}
```

```python
areaWithType_dict
```

```python
df['areaWithType']=df['areaWithType'].map(areaWithType_dict)
```

```python
df.dtypes
```

```python
cat_columns = df.select_dtypes(include='object').columns

for col in cat_columns:
    unique_vals = df[col].unique()
    mapping_dict = {}
    i=1
    for val in unique_vals:
        mapping_dict[val]=i
        i+=1
    df[col] = df[col].map(mapping_dict)
```
