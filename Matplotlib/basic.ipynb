<h4>Scatter</h4>

```python
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,2,10])
plt.scatter(x,y)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,2,10])
plt.scatter(x,y,c="red")
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,2,10])
c=np.array(['red','blue','green','yellow','pink'])
plt.scatter(x,y,c=c)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,2,10])
color=np.array(['red','blue','green','yellow','pink'])
size=np.array([50,200,100,300,40])
plt.scatter(x,y,c=color,s=size)
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
plt.scatter(x, y,alpha=0.5, s = sizes, c = c, cmap='virdis') # cmap = color map, viridis, nipy_spectral, coolwarm  >>>>>here alpha is overlapping point and cmap shows the colormap
plt.colorbar() #this is used to show the colorbar
plt.show()
```

<h4>Pie Chart</h4>

```python
x=np.array([1,2,3,4])
plt.pie(x)
plt.show()
```

```python
x=np.array([1,2,3,4])
label=np.array(["O2","Co2","N2","other"])
plt.pie(x, labels=label,startangle=90)
plt.show()
```

```python
x=np.array([1,2,3,4])
label=np.array(["O2","Co2","N2","other"])
myexplode=[0.2,0,0,0]
plt.pie(x, labels=label,startangle=90, explode=myexplode)
plt.show()
```

```python
x=np.array([1,2,3,4])
label=np.array(["O2","Co2","N2","other"])
myexplode=[0.2,0,0,0]
plt.pie(x, labels=label,startangle=90, explode=myexplode, shadow=True)
plt.show()
```

```python
x=np.array([1,2,3,4])
label=np.array(["O2","Co2","N2","other"])
color=np.array(["navy","cyan","purple","pink"])
myexplode=[0.2,0,0,0]
plt.pie(x, labels=label,startangle=90, explode=myexplode, shadow=True, colors=color)
plt.show()
```

```python
gases=["Nitrogen","Oxygen","Argon","Carbon dioxide","Other Gases"]
percentages=[78.09,20.95,0.93,0.04,0.01]
color=["cyan","pink","yellow","skyblue","lightcoral"]
myexplode=[0,0.1,0,0,0]
plt.figure(figsize=(8,8))
plt.pie(percentages, labels=gases,startangle=90, explode=myexplode, shadow=True, colors=color,autopct='%1.2f%%')
plt.show()
```
